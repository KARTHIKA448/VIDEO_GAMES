###### How the month affected the number of game released?

```python
month_game = df.groupby("Month")["Title"].count().reset_index().sort_values(by="Title",ascending=False)
month_game.columns = ["month","num_of_games"]
```

```python
month_game
```

```python
# reset_index() : used to reset the index of a DataFrame.
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = month_game, x = "month" , y = "num_of_games")
plt.title("MONTH VS NUM_OF_GAMES")
plt.xlabel("MONTH")
plt.ylabel("NUMBER OF GAMES")
plt.grid()
plt.show()
```

    take away:
        from this plot we can understand that:
            The numer of relased games arrives the peak(127) on June probably because the summmer holiday is coming. On December, there were under 50 games released. Hence, summer holiday is the critical moment for the publishers and students are main customers purchasing these games. The numer of released games achieves the second peak on July(112) probably which is also the peak in the second half year. The probably reason could be that people might already get tired of games bought during summer holidays.

```python
df.loc[100,'Platform(s)'].split(',')
```

```python
df["Platform(s)"].nunique()
```

```python
platforms = {} # creates an empty dictionary named platforms to store the counts of each platform.

for index, row in df.iterrows(): # loop iterates through each row of the df
    platform_list = row['Platform(s)'].split(', ') # splits the content of the 'Platform(s)' column into a list of platforms
    for platform in platform_list: # This inner loop iterates through each platform in the platform_list
        if platform in platforms: #  checks if the platform is already a key in the platforms dictionary.
            platforms[platform] += 1 # increments the count of that platform
        else:
            platforms[platform] = 1 #  initializes the count for that platform to 1.
```

```python
platforms
```

###### Platforms on which each game is available

```python
platforms_df = pd.DataFrame([platforms]).T.reset_index()
platforms_df.columns =['platforms', 'num_of_games']
platforms_df
```

###### GENRES

```python
df["Genre(s)"].nunique()
```

```python
df["Genre(s)"].unique()
```

```python
genres = {}

for index, row in df.iterrows():
    genres_list = row['Genre(s)'].split(', ')
    for genre in genres_list:
        if genre in genres:
            genres[genre] += 1
        else:
            genres[genre] = 1
```

```python
genres
```

```python
genre_df = pd.DataFrame(genres,index=[0]).T.reset_index()
genre_df.columns = ["genres","num_of_games"]
genre_df = genre_df.sort_values(by="num_of_games",ascending=False)
```

```python
genre_df
```

```python
plt.figure(figsize = (12,8))
plt.pie(genre_df["num_of_games"],labels = genre_df["genres"],autopct = "%.2f%%")
plt.title("GENRES VS NUMBER OF GAMES")
plt.show()
```

```python
# for knowing the maximum type of genres
genre_df =genre_df.sort_values('num_of_games', ascending=False).head(10)
```

```python
plt.figure(figsize = (12,8))
plt.pie(genre_df["num_of_games"],labels = genre_df["genres"],autopct = "%.2f%%")
plt.title("GENRES VS NUMBER OF GAMES")
plt.show()
```

###### DEVELOPERS

```python
df["Developer(s)"].nunique()
```

```python
df["Developer(s)"].unique()
```

```python
developers = {}

for index, row in df.iterrows():
    developers_list = row['Developer(s)'].split(', ')
    for developer in developers_list:
        if developer in developers:
            developers[developer] += 1
        else:
            developers[developer] = 1
```

```python
developer_df = pd.DataFrame(developers,index=[0]).T.reset_index()
developer_df.columns = ["developers","num_of_games"]
developer_df = developer_df.sort_values(by="num_of_games",ascending=False).head(15)
```

```python
plt.figure(figsize = (12,8))
plt.pie(developer_df["num_of_games"],labels = developer_df["developers"],autopct = "%.2f%%")
plt.title("DEVELOPERS VS NUMBER OF GAMES")
plt.show()
```
